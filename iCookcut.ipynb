```python
import pandas as pd
df = pd.read_csv("iCook_translated.csv", encoding="utf-8")
df = df[["Cuisine","Ingredients"]]
```

```python
def remove_amount(origin_row_data):
    return ''.join([i + (';' if [i.split(':')[0] for i in origin_row_data.replace('}{', ' ').replace('{', '').replace('}', '').split(' ')].index(i) < len([i.split(':')[0] for i in origin_row_data.replace('}{', ' ').replace('{', '').replace('}', '').split(' ')])-1 else '') for i in [i.split(':')[0] for i in origin_row_data.replace('}{', ' ').replace('{', '').replace('}', '').split(' ')]]).replace(' ', '').replace('【', '').replace('】', '').replace('（', '').replace('）', '').replace('(', '').replace(')', '').replace('/', '').replace('-', '').replace('[', '').replace(']', '').replace('／', '').replace(';;', '').replace(';;;', '').replace('《', '').replace('》', '')
```

```python
df['Ingredients'] = df['Ingredients'].apply(remove_amount)
```

```python
import re
for num, i in enumerate(df['Ingredients']):
    #tmpStr = re.sub("[A-Za-z0-9(){}《》［］＜＞﹝﹞「」︰:：％•●?=※\s+\.\!\/_$%^*(+\"\']+|[+——！，？、~～～@#￥%……&*（）。,「」-]+\s","", i)
    tmpStr = re.sub('[^\u4e00-\u9fff]+', '', i)
    df["Ingredients"][num]  = tmpStr 
df    
```

```python
import jieba
import jieba.analyse
jieba.set_dictionary('user_dict.txt')

#建立停用字詞 並過濾
def stopwordslist(filepath):  
    stopwords = [line.strip() for line in open(filepath, 'r', encoding='utf-8').readlines()]  
    return stopwords  

def seg_sentence(sentence):  
    sentence_seged = jieba.cut(sentence.strip())  
    stopwords = stopwordslist('stop_words.txt')
    outstr = ''  
    for word in sentence_seged:  
        if word not in stopwords:  
            if word != '/t':  
                outstr += word  
                outstr += " "  
    return outstr  

df["Ingredients"] = df["Ingredients"].apply(seg_sentence)
df
```

```python
ingredient_row_data = df["Ingredients"] 

each_ingredient_amount_dict = {}

for num, each_row in enumerate(ingredient_row_data):
    try:
        for each_ingredient in each_row.split(' '):
            if each_ingredient in each_ingredient_amount_dict:
                each_ingredient_amount_dict[each_ingredient] += 1
            else:
                each_ingredient_amount_dict[each_ingredient] = 1
    except AttributeError:
        print(num, 'AttributeError')
        #ingredient_row_data.drop([num])
        continue

each_ingredient_amount_dict
```
